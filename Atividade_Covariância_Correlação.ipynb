Carregando os dados e traduzindo as colunas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('StudentPerformanceFactors.csv')

# Tradução das colunas

df = df.rename(columns={
    'Hours_Studied':'Horas_Estudo',
    'Attendance':'Frequencia',
    'Previous_Scores':'Notas_Anteriores',
    'Tutoring_Sessions':'Monitorias',
    'Sleep_Hours':'Horas_Sono',
    'Physical_Activity':'Atividade_Fisica',
    'Exam_Score':'Nota_Final',
    'Gender':'Genero',
    'Parental_Involvement':'Envolvimento_Pais',
    'Access_to_Resources':'Acesso_Recursos',
    'Extracurricular_Activities':'Atividades_Extracurriculares',
    'Motivation_Level':'Nivel_Motivacao',
    'Internet_Access':'Acesso_Internet',
    'Family_Income':'Renda_Familiar',
    'Teacher_Quality':'Qualidade_Professor',
    'School_Type':'Tipo_Escola',
    'Peer_Influence':'Influencia_Colegas',
    'Learning_Disabilities':'Dificuldades_Aprendizagem',
    'Parental_Education_Level':'Escolaridade_Pais',
    'Distance_from_Home':'Distancia_Casa'
})

df.head()
```

Atividade 1 - Covariância e Gráficos de Dispersão

```python
variaveis = [
    'Frequencia',
    'Horas_Estudo',
    'Notas_Anteriores',
    'Monitorias',
    'Horas_Sono',
    'Atividade_Fisica'
]

for coluna in variaveis:

    covariancia = df[coluna].cov(df['Nota_Final'])

    print(f'\nCovariância entre {coluna} e Nota_Final: {covariancia:.4f}')

    plt.figure(figsize=(8,5))

    sns.scatterplot(
        data=df,
        x=coluna,
        y='Nota_Final'
    )

    plt.title(f'{coluna} x Nota Final')
    plt.xlabel(coluna)
    plt.ylabel('Nota Final')

    plt.show()
```

Interpretação

## Frequência x Nota Final

A covariância é positiva.

Interpretação:
Quanto maior a frequência do aluno nas aulas, maior tende a ser sua nota final.

## Horas de Estudo x Nota Final

A covariância é positiva.

Interpretação:
Alunos que dedicam mais horas aos estudos apresentam tendência de obter melhores resultados.

## Notas Anteriores x Nota Final

A covariância é positiva.

Interpretação:
Existe continuidade no desempenho acadêmico dos alunos ao longo do tempo.

## Monitorias x Nota Final

A covariância é positiva.

Interpretação:
Participar de monitorias contribui para melhora do desempenho, embora em menor intensidade.

## Horas de Sono x Nota Final

A covariância é próxima de zero e ligeiramente negativa.

Interpretação:
Não existe evidência de influência significativa das horas de sono sobre a nota final.

## Atividade Física x Nota Final

A covariância é muito próxima de zero.

Interpretação:
A prática de atividade física não apresenta relação linear relevante com o desempenho acadêmico.

Atividade 2 - Heatmap de Correlação

```python
plt.figure(figsize=(12,8))

correlacao = df.select_dtypes(include='number').corr()

sns.heatmap(
    correlacao,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Mapa de Calor das Correlações')
plt.show()
```
