Importação das bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Carregando a base

```python
df = pd.read_csv('StudentPerformanceFactors.csv')

df.head()
```

Visualizando as colunas

```python
print(df.columns.tolist())
```

Qual a distribuição entre homens e mulheres?

```python
distribuicao_genero = df['Gender'].value_counts()

print(distribuicao_genero)
```

Frequência relativa (%)

```python
percentual_genero = df['Gender'].value_counts(normalize=True) * 100

print(percentual_genero)
```

Gráfico

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender')
plt.title('Distribuição por Gênero')
plt.show()
```

Resultado

Gênero	Quantidade
Masculino	3814
Feminino	2793

Percentuais aproximados:

Masculino: 57,73%
Feminino: 42,27%

Interpretação

A base possui maior quantidade de alunos do sexo masculino, representando aproximadamente 58% da amostra.

Qual a média das notas?

```python
media_notas = df['Exam_Score'].mean()

print(f'Média das notas: {media_notas:.2f}')
```

Histograma

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Exam_Score'], kde=True)
plt.title('Distribuição das Notas')
plt.show()
```

Interpretação

A nota média dos alunos é de aproximadamente 67,24 pontos.

Qual a média das horas de estudo?

```python
media_estudo = df['Hours_Studied'].mean()

print(f'Média de horas estudadas: {media_estudo:.2f}')
```

Interpretação

Os alunos estudam em média aproximadamente 20 horas por semana.

Qual a frequência média?

A frequência está representada pela variável Attendance.

```python
media_frequencia = df['Attendance'].mean()

print(f'Frequência média: {media_frequencia:.2f}%')
```

Interpretação

Os alunos apresentam uma frequência média de aproximadamente 80%.

Quantos alunos foram aprovados e reprovados?

Como o dicionário não informa explicitamente uma coluna de aprovação, normalmente considera-se:

```python
df['Status'] = np.where(df['Exam_Score'] >= 60,
                        'Aprovado',
                        'Reprovado')

resultado = df['Status'].value_counts()

print(resultado)
```

Gráfico

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Status')
plt.title('Aprovados e Reprovados')
plt.show()
```

Interpretação

A grande maioria dos alunos foi aprovada, indicando um desempenho geral satisfatório da amostra.
